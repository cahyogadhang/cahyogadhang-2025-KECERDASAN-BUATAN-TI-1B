## Mengatasi Missing Values dengan Drop Data

### 1. Hapus Data yang kosong (NaN)

```python
df = pd.read_csv('./Data.csv')
df.dropna(inplace=True)

df
```

## Mengatasi Missing Values dengan Nilai Baru

### 2. Ganti Data yang kosong (NaN) menjadi :

### <ul><li>Umur = 50 </li><li>Gaji = 25000000</li></ul>

```python
df = pd.read_csv('./Data.csv')
df['Umur'].fillna(50, inplace=True)
df['Gaji'].fillna(25000000, inplace=True)
df
```

## Mengatasi Missing Values dengan Nilai Mean

## 3. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Rata-Rata

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].mean();
mean_gaji = df['Gaji'].mean();

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)
df
```

## Mengatasi Missing Values dengan Nilai Median

## 4. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Median

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].median();
mean_gaji = df['Gaji'].median();

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)

df
```

## Mengatasi Missing Values dengan Nilai Mode

## 5. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Modus

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].mode()[0];
mean_gaji = df['Gaji'].mode()[0];

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)

df
```

## Mengatasi Wrong Data dengan Ubah Data

## 6. Apabila Ada Umur yang diatas 120 maka ubah menjadi 33

```python
df = pd.read_csv('./Data.csv')

for x in df.index:
    if df.loc[x, 'Umur'] > 120:
        df.loc[x, 'Umur'] = 33

df
```

## Soal-1
Buatlah table list Top 5 Customers yang memiliki nilai transaksi paling banyak
---





```python
# Membaca data transaksi dari transactions_1k.csv
transaction_df = pd.read_csv('sample_data/transactions_1k.csv')

# Mengelompokkan berdasarkan CustomerID dan menjumlahkan nilai transaksi (Price)
top_customers = transaction_df.groupby('CustomerID')['Price'].sum().reset_index()

# Mengurutkan berdasarkan total nilai transaksi secara menurun dan mengambil 5 teratas
top_5_customers = top_customers.sort_values(by='Price', ascending=False).head(5)

# Membaca data pelanggan dari customers.csv
customer_df = pd.read_csv('sample_data/customers.csv')

# Menggabungkan dengan data pelanggan untuk mendapatkan informasi Segment dan Currency
top_5_customers = top_5_customers.merge(customer_df, on='CustomerID', how='left')

# Menampilkan tabel hasil
print("Top 5 Customers berdasarkan Nilai Transaksi:")
print(top_5_customers[['CustomerID', 'Segment', 'Currency', 'Price']])

# Opsional: Menyimpan hasil ke file CSV
top_5_customers[['CustomerID', 'Segment', 'Currency', 'Price']].to_csv('top_5_customers.csv', index=False)
```

## Soal-4
Buatlah deskripsi statistik untuk masing masing hari (23,24,25,26) pada dataset transaction_1k
---

```python
# Membaca file CSV ke DataFrame
df = pd.read_csv('sample_data/transactions_1k.csv')

# Mengonversi kolom Amount dan Price ke numerik
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Mengonversi kolom Date ke format datetime
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')

# Mengekstrak hari untuk filtering
df['Day'] = df['Date'].dt.day

# Memfilter data untuk hari 23, 24, 25, dan 26
df_filtered = df[df['Day'].isin([23, 24, 25, 26])]

# Menghitung deskripsi statistik untuk Amount dan Price
stats_amount = df_filtered.groupby(df_filtered['Date'].dt.strftime('%Y-%m-%d'))['Amount'].describe()
stats_sales = df_filtered.groupby(df_filtered['Date'].dt.strftime('%Y-%m-%d'))['Price'].describe()

# Menggabungkan statistik Amount dan Sales
stats_combined = pd.concat([stats_amount, stats_sales], axis=1, keys=['Amount', 'Sales'])

# Menampilkan sebagai DataFrame (tanpa print)
stats_combined
```

## Soal-5
Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi
---

```python
df_transaction['Amount'] = pd.to_numeric(
    df_transaction['Amount'], errors='coerce')
df_transaction['Price'] = pd.to_numeric(
    df_transaction['Price'], errors='coerce')

# Menggabungkan Date dan Time menjadi satu kolom datetime
df_transaction['Datetime'] = pd.to_datetime(
    df_transaction['Date'].astype(
        str) + ' ' + df_transaction['Time'].astype(str),
    format='%Y-%m-%d %H:%M:%S',
    errors='coerce'
)

# Mengekstrak tanggal dan jam
df_transaction['Date'] = df_transaction['Datetime'].dt.strftime('%Y-%m-%d')
# Jam dalam format 0-23
df_transaction['Hour'] = df_transaction['Datetime'].dt.hour

# Mengelompokkan data berdasarkan Date dan Hour, lalu menghitung Amount (sum) dan Sales (sum of Price)
result = df_transaction.groupby(['Date', 'Hour']).agg({
    'Amount': 'sum',  # Jumlah total Amount per jam
    'Price': 'sum'    # Total Price per jam
}).reset_index()

# Menampilkan hasil sebagai tabel DataFrame
result
```

```python
# Konversi kolom Amount dan Price ke numerik
df_transaction['Amount'] = pd.to_numeric(
    df_transaction['Amount'], errors='coerce')
df_transaction['Price'] = pd.to_numeric(
    df_transaction['Price'], errors='coerce')

# Gabungkan kolom Date dan Time menjadi kolom Datetime
df_transaction['Datetime'] = pd.to_datetime(
    df_transaction['Date'].astype(
        str) + ' ' + df_transaction['Time'].astype(str),
    format='%Y-%m-%d %H:%M:%S',
    errors='coerce'
)

# Ekstrak Date dan Hour
df_transaction['Date'] = df_transaction['Datetime'].dt.strftime('%Y-%m-%d')
df_transaction['Hour'] = df_transaction['Datetime'].dt.hour

# Kelompokkan berdasarkan Date dan Hour, lalu hitung total Amount dan Price
grouped = df_transaction.groupby(['Date', 'Hour']).agg({
    'Amount': 'sum',
    'Price': 'sum'
}).reset_index()

# Ambil satu baris per Date dengan Amount terbesar (sesuai contohmu)
daily_max = grouped.loc[grouped.groupby('Date')['Amount'].idxmax()]

# Format ulang dan ubah nama kolom
daily_max = daily_max.rename(columns={'Hour': 'Time', 'Price': 'Gas Sales'})
daily_max['Gas Sales'] = daily_max['Gas Sales'].round(2)

# Jadikan kolom Date sebagai index
daily_max.set_index('Date', inplace=True)

# Tampilkan hasil
daily_max
```
